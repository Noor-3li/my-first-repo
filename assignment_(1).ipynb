# Pandas Assignment: Property Data

```python
import pandas as pd
import numpy as np
```

## Task 1
Read `properties.csv` into a DataFrame named `df` and display the first 5 rows.

```python
# TODO
df = pd.read_csv("/content/properties.csv")
df.head(5)
```

## Task 2
Store the number of missing values in `buyer` as `missing_buyer_count` and whether any missing values exist as `buyer_has_nulls`.

```python
# TODO
missing_buyer_count = df["buyer"].isnull().sum()
buyer_has_nulls = df["buyer"].isnull().any()
print(missing_buyer_count, buyer_has_nulls)
```

## Task 3
Replace missing values in `advertised_price` with the median and store the cleaned DataFrame in `df_filled_price`.

```python
# TODO
df_filled_price = df.copy()
df_filled_price["advertised_price"] = df_filled_price["advertised_price"].fillna(
    df_filled_price["advertised_price"].median()
)
df_filled_price
```

## Task 4
Encode `property_category` so that `Shop=1`, `House=0` and store it in `property_category_encoded`. Missing values should remain missing.

```python
# TODO

mapping = {'Shop': 1, 'House': 0}
df_filled_price["property_category_encoded"] = df_filled_price["property_category"].map(mapping)
df_filled_price[['property_category', 'property_category_encoded']]
```

## Task 5
Create `filtered_df` with rows where location is `Gulberg` or `DHA`, `property_category` is not missing, and `advertised_price > 5000000`. Store the average advertised price in `filtered_avg_price`.

```python
# TODO

filtered_df = df_filled_price[
    (
        (df_filled_price["location"].isin(["Gulberg", "DHA"])) &
        (df_filled_price["property_category"].notnull()) &
        (df_filled_price["advertised_price"] > 5000000)
    )
]

filtered_avg_price = filtered_df["advertised_price"].mean()
print(filtered_avg_price)
filtered_df
```

## Task 6
Count how many times the word `good` appears in the `remarks` of the first property, case-insensitive. Store in `first_good_count`.

```python
# TODO
first_remarks = df_filled_price["remarks"].iloc[0]
first_good_count = str(first_remarks).lower().count("good")
print(first_good_count)
```

## Task 7
Create `discount_amount = advertised_price - sale_price` inside `df_filled_price` and store the average discount in `avg_discount`.

```python
# TODO
df_filled_price["discount_amount"] = (
    df_filled_price["advertised_price"] - df_filled_price["sale_price"]
)
avg_discount = df_filled_price["discount_amount"].mean()
print(avg_discount)
df_filled_price[['advertised_price', 'sale_price', 'discount_amount']].head()
```

## Task 8
Save `filtered_df` to `filtered_properties.csv` without the index.

```python
filtered_df.to_csv("filtered_properties.csv", index=False)
```
